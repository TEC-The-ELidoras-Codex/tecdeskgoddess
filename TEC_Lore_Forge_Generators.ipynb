````python
# 🚀 LIVE WORLD ANVIL API INITIALIZATION
# ===============================
# PRODUCTION MODE - LIVE DEPLOYMENT
# ===============================

import os
import json
import requests
from datetime import datetime

class LiveWorldAnvilAPI:
    """Production World Anvil API client for TEC Lore Forge"""
    
    def __init__(self, force_demo=False):
        self.base_url = "https://www.worldanvil.com/api/boromir"
        
        # Get API credentials from environment variables
        self.api_key = os.getenv('WORLD_ANVIL_API_KEY')
        self.world_id = os.getenv('WORLD_ANVIL_WORLD_ID')
        
        # Force demo mode only if explicitly requested or no credentials
        self.demo_mode = force_demo or not self.api_key
        
        if self.demo_mode:
            print("🔧 DEMO MODE - Testing locally (set WORLD_ANVIL_API_KEY to enable live mode)")
            print("💡 All content generation will work in simulation mode")
        else:
            print("🚀 LIVE API MODE - Production deployment active!")
            print(f"🌍 World ID: {self.world_id}")
            print(f"🔑 API Key: {self.api_key[:20]}...{self.api_key[-10:] if len(self.api_key) > 30 else self.api_key}")
            
        self.headers = {
            "Authorization": f"Bearer {self.api_key}" if self.api_key else "",
            "Content-Type": "application/json", 
            "User-Agent": "TEC-Lore-Forge/2.0 (Production)",
            "Accept": "application/json"
        }
        
        # Initialize local generator storage for demo mode
        self.generators = {}
        
        # Test connection on initialization
        if not self.demo_mode:
            self._test_connection()
    
    def _test_connection(self):
        """Test connection to World Anvil API"""
        print("🧪 Testing World Anvil API connection...")
        try:
            response = requests.get(
                f"{self.base_url}/user",
                headers=self.headers,
                timeout=10
            )
            
            if response.status_code == 200:
                user_data = response.json()
                print(f"✅ Connected successfully!")
                print(f"👤 User: {user_data.get('username', 'Unknown')}")
                return True
            else:
                print(f"❌ Connection failed: {response.status_code}")
                print(f"Response: {response.text}")
                return False
                
        except Exception as e:
            print(f"❌ Connection error: {str(e)}")
            return False
    
    def create_generator(self, generator_data):
        """Create a new generator on World Anvil"""
        
        if self.demo_mode:
            # Store generator data locally for demo mode
            self.generators[generator_data['name']] = generator_data['entries']
            print(f"📝 Demo: Creating generator '{generator_data['name']}'")
            return {"success": True, "demo": True, "data": generator_data}
        
        # Live API call to World Anvil
        url = f"{self.base_url}/world/{self.world_id}/randomgenerators"
        
        # Construct the payload for World Anvil API
        payload = {
            "name": generator_data['name'],
            "description": generator_data['description'],
            "type": "string",  # Generator type
            "entries": generator_data['entries']
        }
        
        try:
            print(f"🌐 Creating live generator: {generator_data['name']}")
            response = requests.post(url, json=payload, headers=self.headers, timeout=15)
            
            if response.status_code in [200, 201]:
                result = response.json()
                print(f"✅ Successfully created generator: {generator_data['name']}")
                return {"success": True, "live": True, "data": result}
            else:
                print(f"❌ Failed to create generator: {response.status_code}")
                print(f"Response: {response.text}")
                return {"success": False, "error": response.text}
                
        except Exception as e:
            print(f"❌ Error creating generator: {str(e)}")
            return {"success": False, "error": str(e)}
    
    def generate_from_table(self, generator_name):
        """Generate content from a specific generator"""
        
        if self.demo_mode:
            # Demo mode generation
            if generator_name in self.generators:
                import random
                result = random.choice(self.generators[generator_name])
                return {"success": True, "result": result}
            else:
                return {"success": False, "error": f"Generator '{generator_name}' not found in demo mode"}
        
        # Live API generation would go here
        # This would call the World Anvil API to generate content
        print(f"🎲 Live generation from: {generator_name}")
        return {"success": False, "error": "Live generation not implemented yet"}

# Initialize the live API
print("🚀 TEC LORE FORGE - LIVE DEPLOYMENT INITIALIZATION")
print("==================================================")

# Check for API credentials
api_key = os.getenv('WORLD_ANVIL_API_KEY')
world_id = os.getenv('WORLD_ANVIL_WORLD_ID')

if api_key and world_id:
    print("🔑 API credentials found - initializing LIVE mode")
    live_api = LiveWorldAnvilAPI(force_demo=False)
else:
    print("⚠️  No API credentials found - running in DEMO mode")
    print("💡 Set WORLD_ANVIL_API_KEY and WORLD_ANVIL_WORLD_ID environment variables for live deployment")
    live_api = LiveWorldAnvilAPI(force_demo=True)

print("📅 Session Date:", datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print("🌟 TEC Visual Sovereignty Protocol: TEC_CSS_072125_V1")
print("⚔️ Mission Status: LIVE DEPLOYMENT READY")
print("==================================================")
print("✅ Live API initialized and ready for generator creation!")
````
